```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('nyc_weather_2018.csv')
data2 = pd.read_csv('weather_stations.csv')
data = pd.merge(data1, data2, left_on='station',right_on='id')
data = data.drop_duplicates(subset=['station', 'datatype'])
data.head()
```

```python
snow = data.query(
    'datatype == "SNOW" and name.str.contains("NJ US")', engine='python')
snow.elevation.mean()
```

```python
fb = pd.read_csv('fb.csv')
new_fb = fb.assign(pct_change=lambda x: x.volume.pct_change())
new_fb.sort_values(by='pct_change', ascending=False).head(5)
```

```python
data3 = pd.read_csv('earthquakes.csv')
new_data3 = data3[data3['magType'] == 'ml']
new_data3 = pd.cut(new_data3.mag, [0, 1, 2, 3, 4, 5])
new_data3.value_counts()
```

```python
data4 = pd.read_csv('weather_by_station.csv')
data4['month'] = pd.DatetimeIndex(data4.date).month
pd.crosstab(index=data4.station, columns=data4.month, colnames=[
            'month'], values=data4.value, aggfunc=np.mean)
```
