```python
cvd = pd.read_csv('covid19_cases.csv')
cvd.head()
```

```python
cvd['Date'] = pd.to_datetime(cvd.dateRep)
```

```python
cvd.set_index('Date', inplace=True)
```

```python
cvd.sort_index().head()
```

```python
cvd.replace({'United_States_of_America': "US",
             "United_ Kingdom": "UK"}, inplace=True)
```

```python
cvd_ = cvd.loc[cvd['countriesAndTerritories'].isin(
    ["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK",  "USA"])]
cvd_.head()
```

```python
cvd1 = cvd_.pivot(columns="countriesAndTerritories",
                  values="cases").fillna(0)
cvd1.head()
```
